```
import requests
from bs4 import  BeautifulSoup



def query(index=-1):
    
    url="./Library _ Yokogawa Electric Corporation.html"
    file=open(url)
    res=file.read()
    #print(res)
    soup=BeautifulSoup(res,"html.parser")
    urllist=soup.find_all("a", class_="thumbnail-wrapper")
    print("strart")
    indx=0
    for i,_ in enumerate(urllist):
        fname = urllist[i]["href"]
        if fname[-3:]=="pdf":
            print(indx,urllist[i]["href"])
            indx+=1

    print("----------------  END  ------------------------")











```
